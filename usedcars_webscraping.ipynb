```python
#Import the required libraries
import numpy as np
import pandas as pd 
import selenium
from selenium import webdriver
#from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.common.exceptions import NoSuchElementException  

pd.options.display.max_colwidth=500
pd.set_option('display.max_rows',None)
import warnings
warnings.filterwarnings('ignore')
```

```python
#Create empty lists for columns of variables
title=[]
Year=[]
model=[]
fuel=[]
distance=[]
transmission=[]
location=[]
price=[]
owner=[]
```

```python
#Connect to the webdriver
driver=webdriver.Chrome(r"C:\Users\dravi\Downloads\chromedriver_win32\chromedriver.exe")
    
```

```python
#Get the URL
driver.get('https://www.olx.in/cars_c84')
time.sleep(0.5)
```

```python
#Maximise the window
driver.maximize_window()
```

```python
#Enter the location in the in the search bar
search=driver.find_element_by_xpath("//input[@class='_2xvhw']")
search.click()
search.send_keys('Kolkata')
time.sleep(1)
```

```python
#Click on the dropdown to select the location
dropdown=driver.find_element_by_xpath('//*[@id="container"]/header/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div[1]/div/span')
dropdown.click()
time.sleep(1)
    
```

```python
#Click the 'Load more' button till all the results corresponding to the location are displayed
for i in range(100):
    try:
        driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
        driver.find_element_by_xpath("//div[@class='JbJAl']").click()
        time.sleep(5)
    except NoSuchElementException:
        pass

```

```python
#Collect the url s of each of the car and store in the list 
url=driver.find_elements_by_xpath("//li[@class='_31j8e']/a")
prod_url=[i.get_attribute('href') for i in url]
```

```python
len(prod_url)
```

```python
#Run through each of the url, collecting the details of each car from each url
for k in prod_url:
        try:
            driver.get(k)
        except:
            pass
        time.sleep(1)
        try:
            head=driver.find_element_by_xpath("//div[@class='_35xN1']").text
            title.append(head.split('(')[0])
            Year.append(head.split('(')[-1].strip(')'))
            location.append('Kolkata')
        except NoSuchElementException:
            title.append(np.nan)
            Year.append(np.nan)
            location.append(np.nan)
            
        try:
            model.append(driver.find_element_by_xpath("//div[@class='_3tLee']").text)
        except NoSuchElementException :
            model.append(np.nan)
        
        try:
            fuel.append(driver.find_element_by_xpath("//div[@class='aOxkz']/div[1]").text)
        except NoSuchElementException:
            fuel.append(np.nan)
        
        try:
            distance.append(driver.find_element_by_xpath("//div[@class='aOxkz']/div[2]").text)
        except NoSuchElementException:
            distance.append(np.nan)
        
        try:
            transmission.append(driver.find_element_by_xpath("//div[@class='aOxkz']/div[3]").text)
        except NoSuchElementException:
            transmission.append(np.nan)
            
        try:
            owner.append(driver.find_element_by_xpath('//*[@id="container"]/main/div/div/div/div[5]/div[1]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]').text)
        except NoSuchElementException:
            owner.append(np.nan)
            
        try:
            price.append(driver.find_element_by_xpath("//div[@class='_3FkyT']").text.strip('â‚¹'))
        except NoSuchElementException:
            price.append(np.nan)
```

```python
driver.close()
```

```python
Above cells are run for different locations 'Hyderabad','Mumbai','Chennai','Delhi','Kolkata','Bengaluru','Visakhapatnam','Ahmedabad','Pune','Kochi','Mysore'.Tried keeping them in a for loop
there were disruptions.So the cells were run for each of the locations and the data is scraped.
```

```python
#Store the data in a dataframe 'usedcars'
usedcars=pd.DataFrame()
usedcars['Title']=title
usedcars['Year']=Year
usedcars['Model']=model
usedcars['Fuel']=fuel
usedcars['Distance Travelled']=distance
usedcars['Transmission']=transmission
usedcars['Location']=location
usedcars['Owner']=owner
usedcars['Price']=price
```
